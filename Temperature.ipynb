```python
plt.figure(figsize=[15,6])
sns.barplot(x=df['Country'],y=df['Avg_Temperature_degC'],hue=df['Year'])
plt.xticks(rotation=90)
plt.title('Avg Temperature in Country Depend on Year')
```

```python
plt.figure(figsize=[12,7])
sns.scatterplot(x=df['Forest_Area_pct'],y=df['Avg_Temperature_degC'],hue=df['Country'],style=df['Year'])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Avg Temperature in country and year with Forest Area pct')
```
